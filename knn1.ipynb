```python
def dist(x1,x2):
    return  np.sqrt((x1-x2)**2).sum()
    
```

```python
def knn(x_train,query_point,y_train,k=5):
    vals=[]
    for ix in range(x_train.shape[0]):
        v=[dist(query_point,x_train[ix]),y_train[ix]]
        vals.append(v)
    updated_vals=sorted(vals)
    pred_arr=np.array(updated_vals[:k])
    pred_arr=np.unique(pred_arr[:,1],return_counts=True)
    print(pred_arr)
    index=pred_arr[1].argmax()
    return pred_arr[0][index]
            
        
```

#make predictions

```python
pred_label=knn(x_train,x_test[1300],y_train)
print(int(pred_label))
```

```python
#computing accuracy over test data
sucess=0;
for ix in range(x_test.shape[0]):
    pred_label=knn(x_train,x_test[ix],y_train)
    if int(pred_label)==int(y_test[ix]):
        sucess=sucess+1
        
accuracy=(sucess)/float(x_test.shape[0])*100
print(accuracy)
        
```
